```python
london = dsb_branches.iloc[np.where(dsb_branches['Branch']=='London')]
oxford = dsb_branches.iloc[np.where(dsb_branches['Branch']=='Oxford')]
nottingham = dsb_branches.iloc[np.where(dsb_branches['Branch']=='Nottingham')]
newcastle = dsb_branches.iloc[np.where(dsb_branches['Branch']=='Newcastle')]
```

```python
cus_locations['London Distance'] = cus_locations.apply(
    lambda row: round(
               3963 * arccos((sin(radians(row['Address Lat'])) * sin(radians(london['Branch Lat'])))
                             + cos(radians(row['Address Lat'])) * cos(radians(london['Branch Lat']))
                                     * cos(radians(london['Branch Long']) - radians(row['Address Long']))), 2),
    axis=1
)
```

```python
cus_locations['Oxford Distance'] = cus_locations.apply(
    lambda row: round(
               3963 * arccos((sin(radians(row['Address Lat'])) * sin(radians(oxford['Branch Lat'])))
                             + cos(radians(row['Address Lat'])) * cos(radians(oxford['Branch Lat']))
                                     * cos(radians(oxford['Branch Long']) - radians(row['Address Long']))), 2),
    axis=1
)
```

```python
cus_locations['Nottingham Distance'] = cus_locations.apply(
    lambda row: round(
               3963 * arccos((sin(radians(row['Address Lat'])) * sin(radians(nottingham['Branch Lat'])))
                             + cos(radians(row['Address Lat'])) * cos(radians(nottingham['Branch Lat']))
                                     * cos(radians(nottingham['Branch Long']) - radians(row['Address Long']))), 2),
    axis=1
)
```

```python
cus_locations['Newcastle Distance'] = cus_locations.apply(
    lambda row: round(
               3963 * arccos((sin(radians(row['Address Lat'])) * sin(radians(newcastle['Branch Lat'])))
                             + cos(radians(row['Address Lat'])) * cos(radians(newcastle['Branch Lat']))
                                     * cos(radians(newcastle['Branch Long']) - radians(row['Address Long']))), 2),
    axis=1
)
```

```python
cus_locations.head()
```

```python
def nearest_branch(row):
  list_of_distances = [row['London Distance'], row['Oxford Distance'], row['Nottingham Distance'], row['Newcastle Distance']]
  nearest_branch_index = list_of_distances.index(min(list_of_distances))
  if nearest_branch_index==0:
    return 'London'
  elif nearest_branch_index==1:
    return 'Oxford'
  elif nearest_branch_index==2:
    return 'Nottingham'
  elif nearest_branch_index==3:
    return 'Newcastle'
cus_locations['Nearest Branch'] = cus_locations.apply(nearest_branch, axis=1)
```

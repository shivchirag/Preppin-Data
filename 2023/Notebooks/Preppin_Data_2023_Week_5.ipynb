```python
df = pd.read_csv('/content/drive/MyDrive/Preppin Data/Data/PD 2023 Wk 1 Input.csv')
```

```python
df.head()
```

```python
df['Bank'] = df['Transaction Code'].apply(lambda x: x.split('-')[0])
```

```python
df['Transaction Date'] = df['Transaction Date'].apply(lambda x: pd.to_datetime(x).strftime('%B'))
```

```python
df = df.groupby(['Transaction Date','Bank']).agg({'Value':'sum'}).reset_index()
```

```python
df['Bank Rank per Month'] = df.groupby(['Transaction Date'])['Value'].rank(ascending=False, method='dense')
df['Bank Rank per Month'] = df['Bank Rank per Month'].astype('int')
```

```python
avg_transaction_value_per_rank = df.groupby('Bank Rank per Month').agg({'Value':'mean'}).reset_index()
df = pd.merge(df, avg_transaction_value_per_rank, on='Bank Rank per Month')
df.rename(columns={'Value_x':'Value','Value_y':'Avg Transaction Value per Rank'}, inplace=True)
```

```python
avg_rank_per_bank = df.groupby('Bank').agg({'Bank Rank per Month':'mean'}).reset_index()
df = pd.merge(df, avg_rank_per_bank, on = 'Bank')
df.rename(columns={'Bank Rank per Month_x':'Bank Rank per Month', 'Bank Rank per Month_y':'Avg Rank per Bank'}, inplace=True)
```

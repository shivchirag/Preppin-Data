```python
df = pd.read_csv('/content/drive/MyDrive/Preppin Data/Data/PD 2023 Wk 1 Input.csv')
```

```python
df.head()
```

```python
df.columns = ['transaction_code', 'value', 'customer_code', 'online_or_in_person', 'transaction_date']
```

```python
df.transaction_code = df.transaction_code.apply(lambda x: x.split('-')[0])
df.columns = ['bank', 'value', 'customer_code', 'online_or_in_person', 'transaction_date']
```

```python
df['online_or_in_person'] = df.online_or_in_person.apply(lambda x: 'online' if x==1 else 'in-person')
```

```python
df.transaction_date = df.transaction_date.astype('datetime64').apply(lambda x: x.strftime('%A'))
```

```python
df.head()
```

```python
total_values_by_bank = df.groupby('bank').agg({'value':'sum'}).reset_index()
total_values_by_bank
```

```python
total_values_by_bank_dow_transaction_type = df.groupby(['bank','online_or_in_person','transaction_date']).agg({'value':'sum'}).reset_index()
total_values_by_bank_dow_transaction_type.head()
```

```python
total_values_by_bank_customer_code = df.groupby(['bank','customer_code']).agg({'value':'sum'}).reset_index()
total_values_by_bank_customer_code.head()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
df = pd.read_csv('/content/drive/MyDrive/Preppin Data/Data/2024/PD 2024 Wk 1 Input.csv')
```

```python
df.head()
```

```python
# Splitting the flight details column using apply and lambda function to create new cols
df['Date'] = df['Flight Details'].apply(lambda x: x.split('//')[0])
df['Flight Number'] = df['Flight Details'].apply(lambda x: x.split('//')[1])
df['From'] = df['Flight Details'].apply(lambda x: x.split('//')[2].split('-')[0])
df['To'] = df['Flight Details'].apply(lambda x: x.split('//')[2].split('-')[1])
df['Class'] = df['Flight Details'].apply(lambda x: x.split('//')[3])
df['Price'] = df['Flight Details'].apply(lambda x: x.split('//')[4])
```

```python
df.head()
```

```python
# Changing the datatypes of the new cols
df['Date'] = pd.to_datetime(df['Date'])
df['Price'] = pd.to_numeric(df['Price'])
```

```python
# Updating the values of Flow Card? column
flow_card_dict = {1:'Yes', 0:'No'}
df['Flow Card?'] = df['Flow Card?'].map(flow_card_dict)

# Dropping the Flight Details col
df = df.drop(columns=['Flight Details'])
```

```python
# Filtering based on the value of Flow Card? col and creating two new dataframes
flow_card_holders = df[df['Flow Card?']=='Yes']
non_flow_card_holders = df[df['Flow Card?']=='No']
```

```python
# Exporting the output
flow_card_holders.to_csv('Flow Card Holders.csv')
non_flow_card_holders.to_csv('Non Flow Card Holders.csv')
```

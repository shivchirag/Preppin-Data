```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
input_file = pd.ExcelFile('/content/drive/MyDrive/Preppin Data/Data/2024/PD 2024 Wk 4 Input.xlsx')
```

```python
dfs = {}
for sheet in input_file.sheet_names:
  dfs[sheet] = pd.read_excel(input_file, sheet_name=sheet)

dfs.keys()
```

```python
flow_card = dfs['Flow Card']
flow_card['Flow Card?'] = 'Yes'

non_flow_card = dfs['Non_flow Card']
non_flow_card['Flow Card?'] = 'No'

non_flow_card2 = dfs['Non_flow Card2']
non_flow_card2['Flow Card?'] = 'No'

seat_plan = dfs['Seat Plan']
```

```python
union_df = pd.concat([flow_card, non_flow_card, non_flow_card2], axis=0)
```

```python
agg_df = union_df.groupby(['Flow Card?', 'Class', 'Seat', 'Row'])['CustomerID'].count().reset_index().rename(columns={"CustomerID":"No. of Bookings"})
```

```python
agg_df.head()
seat_plan.head()
```

```python
joined_df = pd.merge(agg_df, seat_plan, how='right', on=['Class', 'Seat', 'Row'])
```

```python
output_df = joined_df[joined_df['No. of Bookings'].isnull()]
output_df = output_df.drop(columns=['No. of Bookings', 'Flow Card?'])
```
